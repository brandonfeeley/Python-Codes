```python
#import files and csv

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os


df=pd.read_csv(r'C:\Users\Feele\Documents\Football Analysis\Excel\WyScout\Top 5 EU Leagues\Forwards.csv')
```

```python
df.columns
```

```python
df1 = df[['xA per 90','Assists per 90','Key passes per 90','Crosses per 90']] #subset of four random metrics
```

```python
df1.corr()
```

```python
plt.matshow(df1.corr(), cmap="YlGnBu") # darker the colour, the higher the correlation is
plt.show()
```

```python
f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df1.corr(), cmap="YlGnBu", fignum=f.number) # type of plot
plt.xticks(range(df1.shape[1]), df1.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(df1.shape[1]), df1.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```
