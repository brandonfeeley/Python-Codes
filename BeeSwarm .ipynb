```python
fig, ax = plt.subplots(figsize=(10,5))
fig.set_facecolor(background)
ax.patch.set_facecolor(background)

#set up our base layer
mpl.rcParams['xtick.color'] = 'white'
mpl.rcParams['ytick.color'] = 'white'#text_color

ax.grid(ls='dotted',lw=.5,color='white',axis='y',zorder=1)
spines = ['top','bottom','left','right']
for x in spines:
    if x in spines:
        ax.spines[x].set_visible(False)
 
sns.swarmplot(x='Crosses',data=df,color='white',zorder=1)

#plot player 
plt.scatter(x=4.58,y=0,c='red',edgecolor='white',s=200,zorder=2)
#plt.text(s='Name',x=9.87,y=-.04,c=text_color)

#plot average
plt.scatter(x=2.3,y=0,c='#FF00E4',edgecolor='white',s=200,zorder=2)

#title and Labels
plt.title('Crosses',c='white',font = 'consolas', fontsize=18)
plt.xlabel('Dribbles',c=background)# we have a title so need for a xlabel.



#plt.savefig('swarm.png',dpi=500,bbox_inches = 'tight',facecolor=background)
```
